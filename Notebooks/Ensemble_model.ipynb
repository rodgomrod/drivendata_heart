```python
import pandas as pd
import numpy as np
from sklearn.externals import joblib
import gc
```

```python
submission_list = [
    '../submissions/lgbm_1.csv',
    '../submissions/xgb_1.csv',
    '../submissions/catb_1.csv',
#     '../submissions/keras_4.csv'
]
```

```python
list_ = list()
main = pd.read_csv(submission_list[0])
list_.append(main)
for file_ in submission_list[1:]:
    df = pd.read_csv(file_, usecols=['heart_disease_present'])
    list_.append(df)

solutions = pd.concat(list_, axis = 1, ignore_index = True)
```

```python
solutions.head(5)
```

```python
df_sol = solutions.iloc[:,list(range(1, len(submission_list)+1))]
```

```python
df_sol['avg'] = solutions.mean(axis=1)
```

```python
df_sol[:20]
```

```python
final_df = pd.DataFrame({'patient_id': solutions[0], 'heart_disease_present': df_sol['avg']})
```

```python
final_df.to_csv('../submissions/ensemble_5.csv', index=None)
```
